# Исследование объявлений о продаже квартир

В вашем распоряжении данные сервиса Яндекс.Недвижимость — архив объявлений о продаже квартир в Санкт-Петербурге и соседних населённых пунктах за несколько лет. Нужно научиться определять рыночную стоимость объектов недвижимости. Ваша задача — установить параметры. Это позволит построить автоматизированную систему: она отследит аномалии и мошенническую деятельность. 

По каждой квартире на продажу доступны два вида данных. Первые вписаны пользователем, вторые получены автоматически на основе картографических данных. Например, расстояние до центра, аэропорта, ближайшего парка и водоёма.

Датасет содержит 22 столбца. В нем есть данные заполненные автоматически (расстояния до центра, аэропорта, парка или водоема) и данные заполненные вручную. Количество значений в столбцах различаются, соответственно есть пропущенные значения. Надо разобраться какого вида пропуски, что можно удалить, а где их нужно скорректировать или заполнить. В датасете есть данные такие как квартира-студия, в которых отсутсвует кухня или незаполненные данные по наличию парков и водоемов, нужно проанализировать как эти данные влияют на продажу и нужно ли их заполнять. Пропущенные значения в описании саммой недвижимости можно заполнить медианными данными.
Предварительно можно утверждать, что данных достаточно для определения параметров оценки недвижимости. 

## Предобработка данных

Будем поочередно проверять все столбцы на пропуски и корректность данных. При необходимости исправлять или заполнять

Столбец first_day_exposition имеет формат строки, в нем незаполненное время. Приведем его к формату даты и уберем время для наглядности. 

```python
data['first_day_exposition'] = pd.to_datetime(data['first_day_exposition'], format = '%Y-%m-%d')
```

Столбец is_apartment указывает, что недвижимость относится к типу "аппартаменты" и имеет булев формат. Предположительно пропуски связаны с тем, что только собственники апартаментов указывают это при продаже. Соответственно мы можем их исправить на значение False.

```python
data['is_apartment'] = data['is_apartment'].fillna(False)
```

```python
data['is_apartment'].value_counts()
```

```python
data['ceiling_height'].value_counts()
```

```python
data['ceiling_height'].isna().sum()
```

```python
data.loc[data['ceiling_height'].isna(), 'ceiling_height'] = data['ceiling_height'].median()
```

```python
data['ceiling_height'].isna().sum()
```

Проверяем столбец с балконами, в нем большое количество пропусков. Предположительно те, у кото нет балкона просто пропускали это значение при заполнение характеристик недвижимости. Поменяем пропуски на 0 и тип данных на целое число.

```python
data['balcony'].isna().sum()
```

```python
data['balcony'] = data['balcony'].fillna(0)
data['balcony'] = data['balcony'].astype('int')
```

```python
data['balcony'].isna().sum()
```

В столбце площадь кухни пропущенные значения двух видов. Первые это квартиры-студии у них пропуски поменяли на значения 0, так как кухни фактически там нет, во втором случае отсутсвие значений связано скорее всего с пропуском при заполнении данных эти пропуски поменяли на медианное значение.

```python
data.loc[data['studio'] == 1, 'kitchen_area'] = 0
```

```python
data.loc[data['kitchen_area'].isna(), 'kitchen_area'] = data['kitchen_area'].median()
```

```python
data['kitchen_area'].isna().sum()
```

Пропуски в столбце сколько дней было размещено объявление могут свидетельствовать об отсутствии данных о снятие объявления. Фактически мы не знаем продана ли была недвижимость или возможно дата продажи была позднее, чем выгружены данные. 

```python
data['days_exposition'].isna().sum()
```

```python
data['days_exposition'].value_counts()
```

Пропуски в столбце общая площадь заменяем медианными значениями от количества комнат в объекте.

```python
for liv in data['rooms'].unique():
    median = data.loc[data['rooms'] == liv, 'living_area'].median()
    print(liv, median)
    data.loc[(data['living_area'].isna()) & (data['rooms'] == liv), 'living_area'] = median
```

```python
data['living_area'].isna().sum()
```

Удалим пропуски в столбцах населенный пункт и общее количество этажей, доля пропусков там менее 1%. При отсутствии заполненного местонахождения некорректно выполнять дальнейший анализ

```python
data = data.dropna(subset=['locality_name','floors_total'])
```

Изменим тип столбца с ценой для более удобного просмотра

```python
data['last_price'] = data['last_price'].astype('int')
```

В столбце название населенного пункта очень много дубликатов в названиях. Для того чтобы упростить поиск и анализ по месту нахождения объекта уберем тип населенного пункта, оставив только название.

```python
data['locality_name'].unique()
```

```python
data['locality_name'].value_counts()
```

```python
for name in ['деревня', 'поселок','посёлок', 'село', 'коттеджный', 'при железнодорожной', 'Пансионат','садоводческое некоммерческое товарищество', 'станции', 'городской', 'посёлок городского типа', 'городского типа', 'садовое товарищество', 'посёлок городского типа', 'поселок городского типа']:
    data['locality_name'] = data['locality_name'].str.replace(name, '')

                  
```

## Расчёты и добавление результатов в таблицу

Добавляем в таблицу цену квадратногоь метра

```python
data['price_square_meter'] = data['last_price'] / data['total_area']
```

Добавляем столбцы день недели, месяц и год публикации объявления

```python
data['weekday_exposition'] = data['first_day_exposition'].dt.weekday

data['month_exposition'] = data['first_day_exposition'].dt.month

data['year_exposition'] = data['first_day_exposition'].dt.year
```

Делаем категоризацию по этажам: первый, последний, другой.

```python
def floor_groups(row):
    if row['floor'] == 1:
        return 'первый'
    elif row['floor'] == row['floors_total']:
        return 'последний'
    return 'другой'
data['floor_groups'] = data.apply(floor_groups, axis=1)
```

Добавляем столбцы соотношение жилой и общей площади и отношение площади кухни к общей.

```python
data['useful_area_ratio'] = data['living_area'] / data['total_area']
```

```python
data['kitchen_area_ratio'] = data['kitchen_area'] / data['total_area']
```

```python
data.sample(20)
```
