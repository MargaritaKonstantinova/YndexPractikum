# Выявить определяющие успешность игры закономерности.

Доступны исторические данные о продажах игр, оценки пользователей и экспертов, жанры и платформы. Нужно выявить определяющие успешность игры закономерности. Для того чтобы сделать ставку на потенциально популярный продукт и спланировать рекламные компании. 

## Изучение данных из файла

```python
import pandas as pd
import numpy as np
import scipy.stats as sps
import matplotlib.pyplot as plt
import seaborn as sns
```

### Изучение данных из файла 

```python
games = pd.read_csv('/datasets/games.csv', sep = ',')
games.sample(10)
```

```python
games.describe()
```

```python
games.info()
```

#### Вывод

Описание данных
 - name - название игры
 - platform - платформа
 - year_of_release - год выпуска
 - genre - жанр игры
 - na_sales - продажи в Северной Америке (миллионы проданных копий)
 - eu_sales - продажи в Европе (миллионы проданных копий) 
 - jp_sales - продажи в Японии (миллионы проданных копий)
 - other_sales - продажи в других странах (миллионы проданных копий)
 - critic_score - оценка критиков (максимум 100)
 - user_score - оценка пользователей (максимум 10)
 - rating - рейтинг от организации ESRB (англ.Entertainment Software Rating Board). Эта ассоциация определяет рейтинг компьютерных игр и присваивает им подходящию возрастную категорию.
    1. EC («Early childhood») — «Для детей младшего возраста»
    2. E («Everyone») — «Для всех»
    3. K-A ("Kids to Adults") — вариация категории «Для всех»
    4. E10+ («Everyone 10 and older») — «Для всех от 10 лет и старше»
    5. T («Teen») — «Подросткам»
    6. M («Mature») — «Для взрослых 17+»
    7. AO («Adults Only 18+») — «Только для взрослых»
    8. RP («Rating Pending») — «Рейтинг ожидается»
 

Данные за 2016 год могут быть не полными.

### Вывод

В каждой строке информация по игре: ее оценки, год релиза, платформа. В этом столбце всего 2 пропуска, чистоту данных проверем на следующем этапе. На первом этапе нужно привести названию столбцов к нижнем регистру. Поменять тип данных в столбце год, а также проверить почему в столбцах оценка пользователей и рейтинг такой тип данных, при необходимости сменить. Нужно понять почему в некоторых столбцах столько пропусков можно ли из заполнить и чем.

## Предобработка данных 

Будем поочередно проверять все столбцы на пропуски и корректность данных. При необходимости исправлять или заполнять

```python
games = games.rename(columns = {'Year_of_Release':'year_release'})
```

```python
games.columns = map(str.lower, games.columns)
```

Привели все названия столбцов к нижнему регистру и упростили название столбца год релиза.

```python
games.isna().sum()
```

```python
games.duplicated().sum()
```

```python
games.query('name.isnull()')
```

```python
games.dropna(subset = ['name'], inplace = True)
```

В столбце название игры 2 пропуска. Оба наблюдения еще 1993 года и имеет другие пропуски. Удалим их, на дальнейшее исследование они не повлияют.

```python
games['platform'].unique()
```

```python
games['year_release'] = games['year_release'].fillna(2100).astype(int)
```

Поменяем пропуски в столбце год релиза на 2100 и приведем столбец к тип целых чисел. Заполнять этот столбец неизвестными данными или средними не корректно, а для дальнейшего анализа эти данные нужны, поэтому заполнить несуществующим значением . 

```python
games['critic_score'] = games['critic_score'].fillna(-1)
```

Заменяем пропуски на значение -1, отсутствие оценки критиков будет наглядно, но позволит продолжить исследование. 

```python
games['user_score'] = games['user_score'].fillna(-1)
```

```python
games['user_score'] = games['user_score'].replace('tbd', -1)
```

```python
games['user_score'] = pd.to_numeric(games['user_score'])
```

Заменяем пропуски в столбце также значением -1, осутствие оценки будет наглядно, но позволит продолжить исследование. Меняем на числовой тип данных.  

```python
games['rating'].unique()
```

K-A в 1998 году был переименован в E

```python
games['rating'] = games['rating'].fillna('out of rating')
```

```python
games['rating'] = games['rating'].replace('K-A', 'E')
```

Столбец рейтинг пропущенные значения заполняем значением "рейтинг отсутсвует", и объединяем задвоенные значения.

```python
games.info()
```

### Добавление столбца суммарные продажи

```python
games['total_sales']  = games['na_sales'] + games['eu_sales'] + games['jp_sales'] + games['other_sales']
```

```python
games.head()
```

### Вывод

Обработали все данные. Проверили на дубликаты, пропуски, тип данных, уникальность. Откорректировали названия столбцов. Пропущенные значения поменяли на -1 и 2100 для того чтобы продолжить анализ данных, так как пропусков много и они могут при удалении повлиять на выводы. Удалили 2 строки. Поменяли в необходимых столбцах тип данных. Добавили столбец суммарные продажи во всех регионах.

### Выпуск игр в разные годы

```python
ax = games[games['year_release'] != 2100]['year_release'].plot(kind='hist', bins=40).legend()
plt.title('Распределение по годам')
plt.show()
```

#### Вывод

Первые релизы в наших данных начинаются с 80 годов. На тот момент были совсем другие платформы и развитие в техническом плане очень сильно поменяло требования к играм. По графику видно, что активный рост начался в 90 годы когда был первый скачок в техническом плане и появились игры в новом качестве. Максимум у нас пришелся на 2008-2010 года, когда было максимальное количество презентаций новых игр. После 2010 года опять начался спад, скорее всего в связи с "однотипностью" игр, который скорее всего продолжится пока не произойдет или нового скачка в техническом плане или на рынок не презентуют какую-то новую категрию игр чтобы привлечь игроков. Для дальнейшей оценке вполне достаточно взять промежуток 2006-2016 года 

### Продажи по платформам

```python
pd.set_option('display.max_columns', 40)
games.pivot_table(index='platform', columns='year_release', values='name', aggfunc='count', fill_value='-')
```

```python
games.query('year_release > 2000').pivot_table(index='platform', columns='year_release', values='name', aggfunc='count', fill_value='-')
```

В этой таблице наглядно видно когда появлялись и прекращали выпускать игры для каждой платформы. Помним что 2100 это столбец с играми, у которых неизвестно дата релиза.
```python
games_new = games.query('2100 > year_release >= 2006')
```

```python
platform_sales = games_new.groupby('platform')['total_sales'].sum().sort_values(ascending=False).reset_index().head(10)
platform_sales.head(10)
```

```python
top_platforms = platform_sales['platform'].to_list()
top_platforms
```

Нашли топ продаж игр по платформам c 2006 года.

```python
platform_sales.plot.bar( x = 'platform', y = 'total_sales', rot= 30)
plt.show()
```

#### Вывод

Проанализировав полученные данные можно сделать вывод, что средний срок жизни платформы 8 лет. На обновление платформ главным образом влияет техническое развитие. Так как если происходит какой-то прорыв, соответственно все это сразу внедряется в индустрии и начинают выпускать новые игры с лучшей графикой, производительность. По годам видно, что первая смена поколений была в районе 2006 года, а вторая в 2012 году. После обновления платформы выпуск игр снижается в среднем в течение 3 лет. После этого игры как правило для устаревшей платформы не выпускается.

Поэтому потенциально прибыльными в будущем может оставить только 2 платформы.

## Проверка гипотез

### Средние пользовательские рейтинги платформ Xbox One и PC

```python
games_xone = games_actual.query('platform == "XOne" and user_score != -1 ')
```

```python
games_pc = games_actual.query('platform == "PC" and user_score != -1 ')
```

 alpha = 0.05 - уровень статистической значимости (вероятность того, что мы признаем различными совокупности, которые на самом деле не различаются.)
 
 для проверки гипетезы будем использовать t-тест
 
 H0 - средние пользовательские рейтинги платформа Xbox One и PC одинаковые.
 
 H1 - средние пользовательские рейтинги платформа Xbox One и PC различаются.

```python
alpha = 0.05
results = sps.ttest_ind(games_xone['user_score'], games_pc['user_score'], equal_var = False)
print('p-значение:', results.pvalue)

if (results.pvalue < alpha):
    print("Отвергаем нулевую гипотезу")
else:
    print("Не получилось отвергнуть нулевую гипотезу")
```

```python
games_xone['user_score'].mean()
```

```python
games_pc['user_score'].mean()
```

#### Вывод

Проверив гипотезу мы подтвердили, что средние пользовательские рейтинги Xbox One и PC одинаковые. Что подтвержается и посчитанным средним значением 6,5 и 6,4.

 для проверки гипетезы будем использовать t-тест
 
 H0 - средние пользовательские рейтинги жанров Action и Sports одинаковые.
 
 H1 - средние пользовательские рейтинги жанров Action и Sports различаются.
